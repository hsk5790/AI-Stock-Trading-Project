```python
from pytrends.request import TrendReq
import csv
```

```python
pytrends = TrendReq(
     hl='ko'
    ,tz=540
)

keywords=['삼성전자']
```

```python
pytrends.build_payload(
      keywords
    , cat=0
    , timeframe='2022-03-30 2022-04-29'
    , geo='KR'
    , gprop='')
```

```python
get_info = pytrends.interest_over_time()
```

```python
del get_info['isPartial']
```

```python
get_info.to_csv('samsung_search_trend.csv', encoding='utf-8')
```

```python
get_info
```

```python
search_data =pytrends.get_historical_interest(
        keywords
      , year_start= 2017
      , month_start= 4
      , day_start = 1
      , hour_start = 0
      , year_end = 2022
      , month_end = 4
      , day_end = 29
      , hour_end= 0
      , geo='KR'
      , gprop=''
)
```

```python
del search_data['isPartial']
```

```python
search_data
```

```python
search_data.to_csv('samsung_search_trend_hourly.csv',encoding='utf-8')
```
