```python
html = urlopen("https://news.google.com/search?q=%EC%82%BC%EC%84%B1%EC%A0%84%EC%9E%90&hl=ko&gl=KR&ceid=KR%3Ako")
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup)
```

```python
news=soup.select('article > h3 >a')
```

```python
news
```

```python
news=soup.find_all('a', class_='DY5T1d RZIKme')
news
```

```python
for idx, item in enumerate(news):
    title = item.get_text() # 기사제목
    link = 'https://news.google.com'+str(item.get('href')) #뉴스경로 
    print( str(idx+1)+" . "+ title , link)
```
