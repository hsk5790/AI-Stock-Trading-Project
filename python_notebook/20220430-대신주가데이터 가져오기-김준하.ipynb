```python
objStockChart= win32com.client.Dispatch("CpSysDib.StockChart")
Marketchart=win32com.client.Dispatch("CpSysDib.MarketEye")
```

```python
objStockChart.SetInputValue(0,'A005930')
objStockChart.SetInputValue(1,ord('1'))

objStockChart.SetInputValue(3,20000101)
objStockChart.SetInputValue(5,[0,1,2,3,4,5,8,9])
objStockChart.SetInputValue(6,ord('D'))
objStockChart.SetInputValue(7,5)
objStockChart.SetInputValue(9,ord('1'))
Marketchart.SetInputValue(0,(8,9,13,14,))
Marketchart.SetInputValue(1,'A005930')
```

```python
ㅈㄷㄱㄱㅈㄷ
day_list =[]
time_list =[]
open_list=[]
high_list=[]
low_list=[]
close_list=[]
vol_list=[]
amount_list=[]
# mado_list=[]
# masu_list=[]
# chongmado_list=[]
# chongmasu_list=[]


while  True:
    objStockChart.BlockRequest()
    Marketchart.BlockRequest()
    cnt=objStockChart.GetHeaderValue(3)
    print(cnt)
    remain_request_count =  objCpCybos.GetLimitRemainCount(1)
    print( remain_request_count)
    if remain_request_count == 0:
            print('남은 요청이 모두 소진되었습니다. 잠시 대기합니다.')
            while True:
                time.sleep(2)
                remain_request_count =  objCpCybos.GetLimitRemainCount(1)
                if  remain_request_count > 0:
                    print('작업을 재개합니다.')
                    break
                print('대기 중...')
    for i in range(cnt) :
        day_list.append(objStockChart.GetDataValue(0,i))
        time_list.append(objStockChart.GetDataValue(1,i))
        open_list.append(objStockChart.GetDataValue(2,i))
        high_list.append(objStockChart.GetDataValue(3,i))
        low_list.append(objStockChart.GetDataValue(4,i))
        close_list.append(objStockChart.GetDataValue(5,i))
        vol_list.append(objStockChart.GetDataValue(6,i))
        amount_list.append(objStockChart.GetDataValue(7,i))
#         mado_list.append(Marketchart.GetDataValue(0,0))
#         masu_list.append(Marketchart.GetDataValue(1,0))
#         chongmado_list.append(Marketchart.GetDataValue(2,0))
#         chongmasu_list.append(Marketchart.GetDataValue(3,0))
        
        
    if objStockChart.Continue == False :
        break

dict1={'day' : day_list, 'time' : time_list, 'open' : open_list, 'high' : high_list, 'low' : low_list, 'close' :close_list,
        'vol': vol_list, 'amount' : amount_list  }
df = pd.DataFrame(dict1,
                   columns =['day','time','open','high','low','close','vol','amount'])
stockcode='tick'
df.to_csv("{}.csv".format(stockcode), index=False)
```
